### Step 3: Start the Secure Agent Backend
We will run the Express backend server (`server.js`) in the background. It will start on port 3001.

```python
import subprocess
import time

# Start the Node.js agent backend in the background
process = subprocess.Popen(
    ["node", "server.js"],
    cwd="code-kaggle/HACATHON-MATHURA",
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE
)

# Give the server a few seconds to start up
time.sleep(3)
print("✅ Secure Agent Backend started on port 3001")
```

### Step 4: Test the AI Agent
Now we will use Python to send a complex prompt to the AI Agent. The agent will process it and optionally use the `execute_bash` tool via MCP before responding.

```python
import requests
import json

url = "http://localhost:3001/api/analyze"
payload = {
    "prompt": "Run a bash command to echo 'Hello Kaggle from Suvidha AI Agent!'",
    "mimeType": "",
    "imageBuffer": ""
}
headers = {
    "Content-Type": "application/json"
}

try:
    response = requests.post(url, json=payload)
    result = response.json()
    print("🤖 AI Agent Response:\n")
    print(result.get("result", "No result"))
except Exception as e:
    print("Error connecting to agent:", e)
```
